```python
#Sum of first n numbers
n = int(input("Give input"))
su = 0 
for i in range(1,n+1):
    su += i 

su1 = int((n*(n+1))/2)
print(su1)

print(su)
    
```

```python
#Factorial
n = int(input("Give input"))
fac = 1

for i in range (1 , n+1):
    fac *= i 
print(fac)
```

```python
#Count digits
n = int(input("Give input: "))
digit = 0 
while(n > 0):
    digit += 1 
    n = n // 10 

print(digit)
    
    
```

```python
#Sum of digits 
n = int(input("Give Input: "))
ans = 0 

while(n > 0):
    x = n % 10 
    ans += x
    n = n// 10 

print(ans)
```

```python
#Check if a n. is palindrome or not

n = int(input("Give input"))
m = n
rev = 0
digi = 0 
while(n > 0) :
    digi = n%10
    rev = rev*10 + digi
    n = n// 10

if m == rev :
    print("The given no. is palindrome")
else :
    print("The given no. is not palindrome")
```
