```python
#  Avg marks per student
students = ["Bhupinder", "Joginder", "Bhuri", "Aasha Ram", "Love Charger" , "Shiva"]
marks = ([98 , 87 , 88] , [76 , 65 , 98] ,[ 98 , 76  , 98]  , [87  ,67 , 98],[  67 , 65 , 89] ,[ 87, 76 , 92 ])

# Average Marks per Student
avg_marks = np.mean(marks, axis=1)
print("\nAverage Marks per Student:")
for i in range(len(students)):
    print(students[i], ":", avg_marks[i])

# Highest and lowest marks
maxi = max(marks)
print("Maximum marks achieved by students: " ,maxi)
mini = min(marks)
print("Minimum marks achieved by students: " , mini)


# Standard Deviation per Student
std_dev = np.std(marks, axis=1)

print("\nStandard Deviation:")
for i in range(len(students)):
    print(f"{students[i]} : {std_dev[i]:.2f}")

# Ranking Students
ranks = np.argsort(-avg_marks)
print("\nStudent Rankings:")

# Top Students (Above Class Average)
class_avg = np.mean(avg_marks)

print("\nClass Average:", class_avg)

print("\nTop Students (Above Class Average):")
for i in range(len(students)):
    if avg_marks[i] > class_avg:
        print(students[i])

# Subject-wise Analysis
subjects = ["Math", "Science", "English"]

print("\nSubject-wise Average:")
subject_avg = np.mean(marks, axis=0)

for i in range(len(subjects)):
    print(subjects[i], ":", subject_avg[i])

# Percentage Calculation

percentages = (np.sum(marks, axis=1) / 300) * 100

print("\nPercentages:")
for i in range(len(students)):
    print(f"{students[i]} : {percentages[i]:.2f}%")

# Step 9: Percentage Classification
print("\nPercentage Classification:")

for i in range(len(students)):
    if percentages[i] >= 90:
        grade = "A+"
    elif percentages[i] >= 80:
        grade = "A"
    elif percentages[i] >= 70:
        grade = "B"
    elif percentages[i] >= 60:
        grade = "C"
    else:
        grade = "Fail"

    print(f"{students[i]} : {grade}")

#  Visualization (Bar Chart)
plt.bar(students, avg_marks)
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Average Marks of Students")
plt.show()
```
