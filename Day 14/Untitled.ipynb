```python
# Remove duplicate roll numbers while keeping the original order

roll_no = [101, 105, 102, 101, 103, 105, 104, 102]
print("Original Roll Numbers:", roll_no)
unique_rolls = []
seen = set()

for roll in roll_no:
    if roll not in seen:
        unique_rolls.append(roll)
        seen.add(roll)


print("After Removing Duplicates:", unique_rolls)

```

```python
# Maintain a simple contact directory supporting search by name

contacts = {}

n = int(input("Enter number of contacts: "))

for i in range(n):
    name = input("Enter Name: ")
    number = input("Enter Phone Number: ")
    contacts[name] = number

search = input("\nEnter name to search: ")

if search in contacts:
    print("Phone Number:", contacts[search])
else:
    print("Contact not found.")
```

```python
# Given marks of students , print names of students scoring above the class average
student = {
    "Shiva": 89,
    "Kunal": 40,
    "Freebo": 87,
    "Shikha": 77
}

average = sum(student.values()) / len(student)

print("Class Average:", average)

print("Students scoring above average:")

for name, marks in student.items():
    if marks > average:
        print(name, "-", marks)
```

```python
# Built a recommendation engine that suggests products purchased by similar customers but not by the current customer


products = {
    "Shiva": {"Laptop", "Mouse", "Keyboard"},
    "Rahul": {"Laptop", "Headphones", "Mouse"},
    "Kunal": {"Laptop", "Monitor", "Keyboard"},
    "Aman": {"Mouse", "Keyboard", "Printer"}
}

current = input("Current Customer: ")
similar = input("Similar Customer: ")

print("Current Customer Products:", products[current])
print("Similar Customer Products:", products[similar])

recommend = products[similar] - products[current]

print("Recommended Products:", recommend)
```

```python
# Create an airline reservation model and detect double booked seats

seats = {
    "A1": "Shiva",
    "A2": "Rahul",
    "A3": "Kunal",
    "B1": "Aman"
}

seat = input("Enter Seat Number: ")
name = input("Enter Passenger Name: ")

if seat in seats:
    print("Seat Already Booked!")
    print("Booked By:", seats[seat])
else:
    seats[seat] = name
    print("Booking Successful!")

print("\nReservation List:")
for seat, passenger in seats.items():
    print(seat, "->", passenger)
    
```

```python
# develop an online coding contest scoreboard supporting rank updates after every submission.
# Online Coding Contest Scoreboard

scoreboard = {
    "Shiva": 450,
    "Rahul": 380,
    "Kunal": 420,
    "Aman": 350
}

print("Initial Scoreboard")
for name, score in sorted(scoreboard.items(), key=lambda x: x[1], reverse=True):
    print(name, "-", score)


name = input("\nEnter contestant name: ")
new_score = int(input("Enter updated score: "))

scoreboard[name] = new_score

print("\nUpdated Scoreboard")

rank = 1
for name, score in sorted(scoreboard.items(), key=lambda x: x[1], reverse=True):
    print(rank, ".", name, "-", score)
    rank += 1
```

```python
# Built a plagerism checker that compares two assignments based on unique words and similarty percentage



assignment1 = input("Enter Assignment 1: ").lower().split()
assignment2 = input("Enter Assignment 2: ").lower().split()

set1 = set(assignment1)
set2 = set(assignment2)

print("Unique Words in Assignment 1:", set1)
print("Unique Words in Assignment 2:", set2)

# Common words
common = set1.intersection(set2)
print("\nCommon Words:", common)

# Similarity Percentage
similarity = (len(common) / len(set1.union(set2))) * 100

print("Similarity Percentage:", round(similarity, 2), "%")
```
