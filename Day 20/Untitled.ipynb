```python
a = np.array([10 , 20  , 30 , 40])
b = np.array([1 , 2 ,3, 4])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
```

```python
# Slicing 
a = np.array([10 , 20 , 30 , 40 , 50])
print(a[1:4])
print(a[:3])
print(a[::-1])
```

```python
a = np.array([10 , 15 , 20 , 25])
print(a)
mask = a>18
print(mask)
print(a[mask])
print(a>18)
```

```python
a = np.array([1 , 2,3, 45, 6, 7, 7 , 8, 89 , 98 , 987])
print(np.sum(a))
print(np.mean(a))
print(np.median(a))
print(np.max(a))
```
