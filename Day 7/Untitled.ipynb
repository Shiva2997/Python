```python
# 8.
def discount(n):
    if n > 500 :
        return (n*4)//100
        
    elif n > 1000 :
        return (n*6)//100

    elif n > 1500 :
        return (n*8)//100
        
    elif n > 2000 :
        return (n*10)//100
        
    else:
        return (n*2)//100

a = int(input("Price of the items purchased: "))

print("After discount the price of all the items is: " , a - discount(a))
```

```python
# 9.
def grading(name , marks_chemistry , marks_physics , marks_maths , marks_english):

    
    print("Student name: " , name)
    
    if(marks_maths <40 ):
        print("Failed in maths")
    elif(marks_chemistry <40 ):
        print("Failed in chemistry")
    elif(marks_physics <40 ):
        print("Failed in Physics")
    elif(marks_english <40 ):
        print("Failed in English")
    else:
        print("Congratulations !!! You have successfully passed in all subjects 🙌🏻 ")

    print("The average score is: " , (marks_chemistry + marks_physics + marks_maths + marks_english)//4)

a = input("Name")
b = int(input("Chemistry Marks: "))
c = int(input("Physics Marks: "))
d = int(input("Mathematics Marks: "))
e = int(input("English Marks: "))

grading(a , b, c , d , e)


```
