```python
# Create a file and write your introduction
file = open("introduction.txt", "w")

file.write("My name is Shiva Gupta.\n")
file.write("I am a B.Tech CSE student.\n")
file.write("I enjoy programming in Python and C++.\n")
file.write("I am interested in Data Science and Web Development.\n")

file.close()

print("Introduction written successfully.")
```

```python
# Read and display file contents

file = open("introduction.txt", "r")

content = file.read()

print(content)

file.close()
```

```python
# Append new content to a file

with open("introduction.txt", "a") as file:
    file.write("\nMy hobby is playing cricket.")
    file.write("\nI love learning Python.")

print("Content appended successfully.")
```

```python
# Count number of lines in a file

with open("introduction.txt", "r") as file:
    lines = file.readlines()

print("Number of Lines:", len(lines))
```

```python
# Count number of words in a file

with open("introduction.txt", "r") as file:
    content = file.read()

words = content.split()

print("Number of Words:", len(words))
```

```python
# Search a word in a file

with open("introduction.txt", "r") as file:
    content = file.read()

word = input("Enter the word to search: ")

if word in content:
    print("Word found.")
else:
    print("Word not found.")
```

```python
# Copy one file into another

with open("source.txt", "r") as source:
    content = source.read()

with open("destination.txt", "w") as destination:
    destination.write(content)

print("File copied successfully.")
```

```python
# Convert file contents into upper case

with open("input.txt", "r") as file:
    content = file.read()

with open("output.txt", "w") as file:
    file.write(content.upper())

print("File converted to uppercase successfully.")
```
