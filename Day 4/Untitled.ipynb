```python
n = int(input("Give input: "))
og = n 
rr = 0
while(n > 0 ):
    digit = n % 10 
    rr = rr * 10 + digit 
    n = n // 10 

if(og == rr):
    print("Palindrome")
else :
    print("Not Palindrome")


    
```
