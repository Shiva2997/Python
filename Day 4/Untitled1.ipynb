```python
#Nested loops
for i in range(1, 4):
    print("Row", i)

    for j in range(1, 4):
        print("    Column", j)

    print()
```

```python
for i in range(1 , 3):
    print('*' * 5)
    for j in range(1):
        print('*' * 5)
```

```python
'''
for i in range(1 , 5):
    print("1 2 3 4")

'''
for i in range(3):
    for j in range(1 , 5):
        print(j, end=" ")
    print()
```

```python
n = 2 
for i in range (3):
    for j in range(3):
        print(n , end=" ")
        n += 2 
    print()
```

```python
n = 1 

for i in range(3):
    for j in range(3):
        print(n , end=" ")
        n += 1 
    print()
```

```python
# Multiplication grid

for i in range(1 , 4):
    for j in range(1 , 4):
        print(i*j , end=" ")
    print()
```
