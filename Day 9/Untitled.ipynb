```python
def countdown(n):
    if n == 0 :
        return 
    print(n)
    countdown(n-1)

countdown(10)
```

```python
def fact(n):
    if n == 1:
        return 1

    return n * fact(n - 1)

print(fact(5))
```

```python
#Recursive Numbers to Words
def conver(n):
    if n == 0 :
        return
        
    dig = n%10 ;
    
    match dig:
        case 1:
            print("One")
        case 2 :
            print("Two")
        case 3 :
            print("Three")
        case 4 :
            print("Four")
        case 5 :
            print("Five")
        case 6 :
            print("Six" )
        case 7 :
            print("Seven")
        case 8 :
            print("Eight")
        case 9 :
            print("Nine")

    conver(n//10)
            
            
conver(78978)
```
