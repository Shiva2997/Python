```python
# A function which is the largest no. out of two numbers

def larg(a , b):
    if a > b:
        return a 
    else :
        return b 

n = int(input())
m = int(input())

print("The largest no. is " ,  larg(n , m))
```

```python
# Even or odd

def e_or_o(a ):
    if(a % 2 == 0 ):
        return "even" 
    else :
        return "odd"

n = int(input("Give input: "))

print("The no. is " ,  e_or_o(n))
    
```

```python
def largest(*n):
    big = 0 
    for i in n:
        if i > big :
            big = i

    return big 

print(largest(100 , 1  , 4 , 5 , 8 , 9 , 67 , 78))
        
        
```
