```python
# Convert a snake case string into camel case
s = input("Enter snake_case string: ")

words = s.split("_")

camel = words[0]

for i in range(1, len(words)):
    camel += words[i].capitalize()

print(camel)
```

```python
# Count the no. of letters digits , spaces and special characters
s = input("Enter a string: ")

letters = 0
digits = 0
spaces = 0
special = 0

for ch in s:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1
    else:
        special += 1

print("Letters:", letters)
print("Digits:", digits)
print("Spaces:", spaces)
print("Special Characters:", special)
```

```python
#Check whether two strings are anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not Anagrams")
```

```python
# Find the longest word in the sentence
sentence = input("Enter a sentence: ")

words = sentence.split()

longest = ""

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
print("Length:", len(longest))

```

```python
# Q1. Netflix Content Dashboard
#Concept: Indexing & Slicing
#Shows list provided . Display first , last , top , last 3 , alternative shwos , reverse using slicing


movies = [
    "Inception",
    "Interstellar",
    "The Dark Knight",
    "Avatar",
    "Joker",
    "Oppenheimer",
    "Dune",
    "The Batman",
    "Tenet",
    "Gladiator"
]

print("Available Movies:")
print(movies)

# First movie
print("\nFirst Movie ->", movies[0])


# First three movies
print("\nTop Picks:", movies[0:3])

# Last three movies
print("\nLast Three:" , movies[-3:])

#Alternative shows
print("\nAlternative Shows:" , movies[::2])
```

```python
#Hospital Patient Montering
#Travelsal
#Heart rate list. Print patients with heart rate > 100 BPM and count them 

bpm_list = [78, 115, 92, 104, 85, 126, 98, 101, 87, 119]

high_bpm = 0

print("Patients with High Heart Rate:\n")

for bpm in bpm_list:
    if bpm > 100:
        print(bpm, "BPM")
        high_bpm += 1

print("\nNumber of Patients with BPM above 100:", high_bpm)
```

```python
#Company payroll analysis
#Conpect: built-in fns
#Salary list. Display highest , lowest , total and average salary

pay_list = [45000, 62000, 38000, 81000, 56000, 72000, 49000, 67000]

print("Monthly Payroll:", pay_list)

print("\nMaximum Salary:", max(pay_list))
print("Minimum Salary:", min(pay_list))
print("Total Payroll:", sum(pay_list))
print("Average Salary:", sum(pay_list) / len(pay_list))
```
