```python
# Q1. Create a Book class wiht attributes title , author , price . Add a method display_info() that prints all details.

class Book:

    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def display_info(self):
        print("Title :", self.title)
        print("Author:", self.author)
        print("Price :", self.price)


book1 = Book("The Alchemist", "Paulo Coelho", 499)

book1.display_info()
```

```python
# Q2 Create a Student class with roll_no , name , marks. Add a methd get_grade() that returns "A" if marks >= 90 , "B" if >= 75 , else "C"


class Student:

    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks

    def get_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"



student1 = Student(101, "Shiva", 97)

print("Roll No:", student1.roll_no)
print("Name:", student1.name)
print("Marks:", student1.marks)
print("Grade:", student1.get_grade())
```

```python
# Q4. Create a Rectangle class with length and width. Add method area() and perimeter()

class Rectangle:
    def __init__(this , l , w):
        this.l = l
        this.w = w

        
    def area(this):
        print("Area of Rectangle is " , this.l*this.w)

    def perimeter(this):
        print("Perimeter is " , this.l*2 + this.w*2)

rect = Rectangle(15 , 25)
rect.area()
rect.perimeter()
```

```python
# Q5. Create an Employee class with name and salary . Add a method give_raise(percent) that updates self.salary in place(no return , just modify and print new salary).

class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def give_raise(self, percent):
        self.salary = self.salary + (self.salary * percent / 100)
        print("Updated Salary:", self.salary)


emp1 = Employee("Shiva", 50000)

print("Name:", emp1.name)
print("Current Salary:", emp1.salary)

emp1.give_raise(10)
```

```python
# Create an Employee class with private attribute __salary . Add a getter and a setteler -- the setter should reject negative salaries , Add a method apply_raise(percent) that increases salary using setter internally



























class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    # Getter
    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, salary):
        if salary < 0:
            print("Salary cannot be negative.")
        else:
            self.__salary = salary

    # Apply Raise
    def apply_raise(self, percent):
        new_salary = self.__salary + (self.__salary * percent / 100)
        self.set_salary(new_salary)
        print("Updated Salary:", self.__salary)


emp = Employee("Shiva", 50000)

print("Current Salary:", emp.get_salary())

emp.apply_raise(10)

emp.set_salary(-2000)

print("Final Salary:", emp.get_salary())
```
