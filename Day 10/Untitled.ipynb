```python
# Check palindrome

check = True

n = input("Give input: ")
l = len(n)
i = 0 
while(i < l//2):
    if n[i] != n[l-i-1] :
        check = False
        break

    i += 1 

if check :
    print("Palindrome")
else :
    print("Sorry bud")
    
```
