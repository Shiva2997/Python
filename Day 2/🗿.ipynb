```python

a = int(input("Enter your no: "))

n = a/2 
i = 2 
isPrime = True
while(i <= n) :
    if(a % i == 0):
        print("Not prime")
        isPrime = False
        break 
        
    i += 1

if(isPrime == True):
    print("Prime")
```

```python

#first user

# 10 % offer on 10th order with a purchase of over 1500

# 4 % offer on purchase of over 500

# 6% offer on purchase of over 1000

order_no = int(input("Enter order number: "))
amount = float(input("Enter purchase amount: "))

discount = 0

if amount > 1500:
    discount = 10
elif amount > 1000:
    discount = 6
elif amount > 500:
    discount = 4

if order_no % 10 == 0:
    discount = max(discount, 6)

discount_amount = amount * discount / 100
final_amount = amount - discount_amount

print("Discount:", discount, "%")
print("Discount Amount:", discount_amount)
print("Final Amount:", final_amount)



```
