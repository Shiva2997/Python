```python
# Create an abstract class vehicle with an abstract method fuel_type(). Create PetrolCar and ElectricCar subclasses , each implementing fuel_type() differently

from abc import ABC, abstractmethod

class Vehicle(ABC):
    
    @abstractmethod
    def fuel_type(self):
        pass

class PetrolCar(Vehicle):
    
    def fuel_type(self):
        return "This car runs on Petrol (Gasoline)."

class ElectricCar(Vehicle):
    
    def fuel_type(self):
        return "This car runs on Electricity (Battery power)."

petrol_car = PetrolCar()
electric_car = ElectricCar()

print(petrol_car.fuel_type())
print(electric_car.fuel_type())
```

```python
# Create a base class Person with name and age . Create Teacher and Student subclasses , each adding one extra attribute (e.g. subject for Teacher , roll_no for Student) and overriding a common method role() to describe who they are.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def role(self):
        return f"I am {self.name}, a person."

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def role(self):
        return f"I am {self.name}, a teacher. I teach {self.subject}."

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

    def role(self):
        return f"I am {self.name}, a student. My roll number is {self.roll_no}."

teacher = Teacher("Alice", 35, "Physics")
student = Student("Bob", 15, "S101")

print(teacher.role())
print(student.role())
```

```python
# Create Sqaure and Triangle classes , each with a draw() method that prints a different simple pattern(e.g. using *) . Put objects of both in list and a loop through , calling draw() on each.

class Square:
    def draw(self):
        print("****")
        print("*  *")
        print("*  *")
        print("****")
        print()

class Triangle:
    def draw(self):
        print("   * ")
        print("  *** ")
        print(" ***** ")
        print("*******  ")
        print()


# Creating objects and putting them in a list
shapes = [Square(), Triangle()]

# Looping through and calling draw() on each object
for shape in shapes:
    shape.draw()
```
