```python
records = [
    ['Aditi' ,  29],
    ['Shruti' , 22],
    ['Aaroo' , 89],
    ['Aknoj' , 54],
    [ "Kitten", 45] 
    
]

df = pd.DataFrame(records , columns=['Name' , 'Age'])
df.isnull()
#df.notnull()
#df.notnull().count()
```

```python
# dropna()
df.dropna()
df.dropna(subset = ['Marks'])
df.dropna(how='all')
```

```python
#fillna()
df['Marks'] = df['Marks'].fillna(0)
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df.fillna(method='ffile') # forward fill from the previous row
```

```python
#replace()
df['Gender'] = df['Gender'].replace({'M': 'Male' , 'F' : 'Female'} ]

# rename
df = df.rename(columns={'StudentID' : 'Student_ID'})                                
```

```python
#duplicate rows
df.duplicated() #boolean mask of duplicate rows
df.drop_duplicate()
df.drop_duplicates(subset=['Name'] , keep='first')
```
